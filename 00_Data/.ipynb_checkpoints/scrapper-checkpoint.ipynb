```python
import requests
import pandas as pd
import numpy as np
import pickle
import json
from tqdm import tqdm

from util_scrape import get_player_data, get_todays_data, save_obj, load_obj

restart_database = False 
check_response   = False
```

```python
# load previously scraped player data
p_names = pd.read_csv('00_global_player_database/player_list_v2.csv').col.tolist()
len(p_names)
```

```python
# respnese check to make sure api call exists for each player - look at the rejects and figure out why
listy = []
if check_response == True:
    for i in p_names:
        url = 'https://api-prod.footballindex.co.uk/v2/timeseries?id='+i+'&period=30d&interval=1d'
        response = requests.get(url)
        if str(response) == '<Response [404]>':
            print(i)
            listy.append(i)
            
```

```python
# get player data!
player_dict = {}

if restart_database == True:
    for nam in p_names:
        name, data = get_player_data(nam)
        # append to player dictionary
        player_dict.update({name:data})

    # save off global player database
    save_obj(player_dict, 'FI_global_player_price_data_full_prem_v0')
    load_obj('FI_global_player_price_data_full_prem_v0')
```

```python
# load player database 
player_database = load_obj('FI_global_player_price_data_full_prem_v5')

# load refreshed player data
player_dict_update = {}
for nam in tqdm(p_names):
    name, data = get_player_data(nam)
    
    if name in player_database:
          # merge with player database
        data = pd.concat((player_database[name], data[data.DateTime >= player_database[name]['DateTime'].max()]), axis=0)
        data.drop_duplicates(subset='DateTime',inplace=True, keep='last')
        data.sort_values('DateTime', inplace=True)
    else:
        data.sort_values('DateTime', inplace=True)
    
    # merge with data for today
    name_check, data_today = get_todays_data(name)
    # merge with data for today
    data = pd.concat((data, data_today), axis=0)
    data.drop_duplicates(subset='DateTime', inplace=True, keep='last')
    data.sort_values('DateTime', inplace=True)
    data.reset_index(inplace=True, drop=True)
  
    # append to player dictionary
    player_dict_update.update({name:data})
    
# save off new player database    
save_obj(player_dict_update, 'FI_global_player_price_data_full_prem_v5')
```

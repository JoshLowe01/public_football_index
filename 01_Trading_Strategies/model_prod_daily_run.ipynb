```python
#Credentials 
server_url_    = ''
api_token     = ''
client = sb.SparkBeyondClient(server_url=server_url_, api_key=api_token)
```

```python
# cosmetics
data_run = df[['Date', 'PlayerName', 'EndofDayPrice', 'MatchdayScore', 'MediaScore',
                     'MatchdayDividends', 'MediaDividends']]
data_run.dropna(inplace=True)

# Add in technical indicators - EWMAC
data_run['ewm_8'] = data_run['EndofDayPrice'].ewm(span=8).mean()
data_run['ewm_16'] = data_run['EndofDayPrice'].ewm(span=16).mean()
data_run['ewm_32'] = data_run['EndofDayPrice'].ewm(span=32).mean()
data_run['ewm_64'] = data_run['EndofDayPrice'].ewm(span=64).mean()
data_run['ewmc_8_32'] = data_run['ewm_8'] - data_run['ewm_32']
data_run['ewmc_16_64'] = data_run['ewm_16'] - data_run['ewm_64']

# subset on time 
data_run =  data_run[data_run.Date > '2019-04-01']
data_run.reset_index(drop=True, inplace=True)

# rename cols for SB
data_run.columns = data_run.columns.str.lower()
data_run.columns.name = 0
```

```python
# Validation on the left out test data
# Call the model
model_ = client.review(project_name = 'FI_Josh_Lowe', revision = 32)

timeseries_data = data_run[['date', 'playername', 'endofdayprice', 'matchdayscore', 'mediascore','matchdaydividends',
                         'mediadividends','ewm_8', 'ewm_16','ewm_32','ewm_64','ewmc_8_32','ewmc_16_64']]

timeseries_context = sb.Contexts.TimeSeriesMap(data=timeseries_data,
                                               time_column = 'date', name = 'day_windows', key_column='playername') 

prediction = model_.predict(data_run[['date', 'playername', 'endofdayprice', 'matchdayscore', 
                                 'mediascore','matchdaydividends', 'mediadividends',
                                 'ewm_8', 'ewm_16','ewm_32','ewm_64','ewmc_8_32','ewmc_16_64']],
                           context_datasets = [timeseries_context])
```

```python
prediction_result =  prediction.data
prediction_result = pd.concat((data_run.reset_index(drop=True), prediction_result.reset_index(drop=True)), axis=1)
prediction_result['target_28d_predicted'] = np.exp(prediction_result['target_28d_predicted']) - 100

prediction_result['date'] = pd.to_datetime(prediction_result['date'])
prediction_result = prediction_result[prediction_result['date']  == prediction_result['date'].max()]
prediction_result.head(5)
```

```python
# Decision toolbox

position_scaled_on       = 'scaled_desicion' # target.lower()+'_predicted' #
position_scaled_on_value = 10 # 2
position_big_on_value    = 15

position_small          = 25
position_big_multiplier = 4

W_ml_model   = 0.5
W_ewmac_f    = 0.3
W_ewmac_s    = 0.2


# scale the ewmac to a % of the price
prediction_result['ewmc_8_32_scaled'] = (100/prediction_result['endofdayprice']) * prediction_result['ewmc_8_32']
prediction_result['ewmc_16_64_scaled'] = (100/prediction_result['endofdayprice']) * prediction_result['ewmc_16_64']

# get scaling factors so everything is scaled to -/+20 with -/+10 being the respective means of the -ve's and +ve's
def scaler(string, scaler_neg, scaler_pos):
    series = prediction_result[string]
    series = np.where(series>0, series*scaler_pos, series*scaler_neg)
#     print(string + ' Negative scaler: %.2f  Positive scaler: %.2f: ' % (scaler_neg, scaler_pos))
    series = series.clip(-20, 20)
    return series

prediction_result['ewmc_8_32_scaled'] = scaler('ewmc_8_32_scaled', 2.25, 1.29)
prediction_result['ewmc_16_64_scaled'] = scaler('ewmc_16_64_scaled', 1.54, 0.75)
prediction_result[target.lower()+'_predicted_scaled'] = scaler(target.lower()+'_predicted', 3.28, 7.48)


prediction_result['scaled_desicion'] = W_ml_model*prediction_result[target.lower()+'_predicted_scaled'] + W_ewmac_s*prediction_result['ewmc_16_64_scaled'] + W_ewmac_f*prediction_result['ewmc_8_32_scaled']

# determine position size
prediction_result['position_lots']   = (position_small/prediction_result['endofdayprice']).astype(int)
prediction_result['position_lots']   = np.where(prediction_result[position_scaled_on] > position_big_on_value, 
                                                prediction_result['position_lots']*position_big_multiplier, 
                                                prediction_result['position_lots'])

# calculate $ returns and various metrics
prediction_result['predicted_value'] = prediction_result['position_lots'] * prediction_result['endofdayprice'] * abs((1+ (prediction_result[target.lower()+'_predicted']/100)))
prediction_result['spend']     = prediction_result['position_lots'] * prediction_result['endofdayprice']

# Take just the players expected to increase in value
long_pos = prediction_result[prediction_result[position_scaled_on] > position_scaled_on_value]

print('| Total Spend: %.0f |' % (long_pos['spend'].sum()))
                                                                            
# look at weekly actions after aggregating
long_pos['num'] = 1
long_pos_grouped = long_pos.groupby('date').agg('sum')
long_pos_grouped['predicted_ROI']       = (100/long_pos_grouped['spend']) * (long_pos_grouped['predicted_value']-long_pos_grouped['spend']) 
display(long_pos_grouped[['spend','predicted_value','predicted_ROI','num']].tail(5))

```
